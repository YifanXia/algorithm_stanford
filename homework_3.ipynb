```python
class Vertex:
    
    def __init__(self, id_, neighbours):
        self._id = id_
        self._neighbours = set(neighbours)
        
    def get_id(self):
        return self._id
    
    def get_neighbours(self):
        return self._neighbours
    
    def __add__(self, other):
        '''
        Merge two vertices
        '''
        return Vertex(min(self.get_id(), other.get_id()), self._neighbours | other._neighbours)
```
